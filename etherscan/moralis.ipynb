```python
params = {
    "chain": "eth",
    "address": ADDRESS,
    "limit": 100,  # 100100
    "page": 1      
}

all_results = []
while True:
    print(f"Requesting page {params['page']}...")

    result = evm_api.token.get_wallet_token_transfers(api_key=api_key, params=params)
    transfers = result.get("result", [])

    # 
    if not transfers:
        print(f"No more data at page {params['page']}. Stopping.")
        break

    all_results.extend(transfers)
    print(f"Page {params['page']} returned {len(transfers)} records.")
    params["page"] += 1

print(f"Total records fetched: {len(all_results)}")
```
